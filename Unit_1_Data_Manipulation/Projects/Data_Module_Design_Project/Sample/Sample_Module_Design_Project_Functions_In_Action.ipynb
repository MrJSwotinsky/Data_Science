# `js_plot` Function 1 In Action:
### `double_bar_graph()`

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import js_plot 

categories = ['Average Q1 Sales','Average Q2 Sales','Average Q3 Sales','Average Q4 Sales']
a_sales = [25000, 28000, 19000, 29000]
b_sales = [7000, 12000, 18000, 32000]

js_plot.double_bar_graph('Company A',a_sales,'Company B',b_sales,categories,'Quarterly Sales',1)
```

# `js_plot` Function 2 In Action:
### `compare_to_mean_graphs()`

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import js_plot 

global_waste = pd.read_csv('Global Waste by Country 2019 (v.sp22) - Global Waste.csv')

records = ['Canada','United States']
columns = ['MSW per Capita (tons/people)','Hazardous Waste per Capita (tons/people)','Human Development Index']

js_plot.compare_to_mean_graphs(global_waste,'Name',records,columns)
```
